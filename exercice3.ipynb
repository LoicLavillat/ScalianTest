```python
import pandas as pd
import json
import matplotlib.pyplot as plt


df1 = pd.read_csv("searches.csv", sep='^',usecols=['Destination','Date'])
row_count = sum(1 for line in open("searches.csv"))
print("number of lines :"+str(row_count))

#extract flights arriving to Airport we want to plot
madrid = df1.loc[df1['Destination'] == 'MAD']
malaga = df1.loc[df1['Destination'] == 'AGP']
barlelona = df1.loc[df1['Destination'] == 'BCN']
 
#we want to group flights monthly, we analyzed the file and found it is only 2013 flights
madrid['Date'] = madrid['Date'].apply(lambda m:str(m).split('-')[1])
malaga['Date'] = malaga['Date'].apply(lambda m:str(m).split('-')[1])
barlelona['Date'] = barlelona['Date'].apply(lambda m:str(m).split('-')[1])

#if several years in file
#madrid['Date'] = madrid['Date'].apply(lambda m:str(m).split('-')[0]+'-'+str(m).split('-')[1]) 
#malaga['Date'] = malaga['Date'].apply(lambda m:str(m).split('-')[0]+'-'+str(m).split('-')[1]) 
#barlelona['Date'] = barlelona['Date'].apply(lambda m:str(m).split('-')[0]+'-'+str(m).split('-')[1])

#aggregation of data
dataMadrid = madrid.groupby('Date').Date.count()
dataMalaga = malaga.groupby('Date').Date.count()
dataBarlelona = barlelona.groupby('Date').Date.count()

plt.plot(dataMadrid.index,dataMadrid.values,dataMalaga.index,dataMalaga.values,dataBarlelona.index,dataBarlelona.values)
plt.legend(('Madrid', 'Malaga','Barcelona'),
           loc='upper right')
plt.ylabel('Number of searches')
plt.xlabel('Month')
plt.show()




  

```
